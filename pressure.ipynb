```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

```python
def get_pressure(height, temp=15, rel_humidity=0):
    """
        Computes atmospherique pressure with the barometric formula
        https://en.wikipedia.org/wiki/Barometric_formula
        temp: in °C
        rel_humidity: Relative humidity %
    """
    P0 = 1013.25 # Sea level, in hPa (standardized)
    L = 0.00976 # Temperature lapse rate, in K/m
    T0 = 288.16 # Sea level standard temperature, in K
    g = 9.80665 # Earth-surface gravitational acceleration, in m/s^2
    M = 0.02896968 # Molar mass of dry air, in kg/mol
    R0 = 8.314462618 # Universal gas constant, in J/(mol·K)
    
    return P0 * (1 - (L * height) / T0) ** ((g * M) / (R0 * L))

# Sample pressures at different altitudes
altitudes = np.arange(0, 10001, 500)
pressures = get_pressure(altitudes)
# TODO: adjust temperature (!) and potentially relative humidity

# Plot
plt.plot(altitudes / 1000, pressures)
plt.xlabel('Altitude (km)')
plt.ylabel('Pressure (hPa)')
plt.show()

# Table
pd.DataFrame(data={
    'Altitude (m)': altitudes.astype(int),
    'Pressure (hPa)': pressures.astype(int)
})
```
